## Load in Data

We'll load in the data set and take a cursory look around.

```python
recent_grads = pd.read_csv("recent-grads.csv")
recent_grads.head(5)
```

```python
recent_grads.info()
```

```python
recent_grads.describe()
```

## Clean up Data
Drop any rows with null values

```python
recent_grads.dropna(inplace=True)
recent_grads.info()
```

Looks like only one row was dropped due to missing values so we should be good  as far as that goes.

## Basic Visualization

```python
recent_grads.plot(x="Sample_size", y="Median", kind="scatter")
```

```python
recent_grads.plot("Sample_size", "Unemployment_rate", kind="scatter")
```

```python
recent_grads.plot("Full_time", "Median", kind="scatter")
```

```python
recent_grads.plot("ShareWomen", "Unemployment_rate", kind="scatter")
```

```python
recent_grads.plot("Men", "Median", kind="scatter")
```

```python
recent_grads.plot("Women", "Median", kind="scatter")
```

### Observations
Non of these revealed any interesting relationships. However because are based on number of respondants it's not surprising that this is the case.

```python
recent_grads["Full_time"].hist(bins=50, range=(0,50000))
```

```python
pd.tools.plotting.scatter_matrix(recent_grads[["Median", "Unemployment_rate"]], figsize=(10,10))
```

```python
recent_grads.sort_values(by="Median", inplace=True, ascending=False)
```
